```python
weight = list(np.array(poids))
seuil = list(np.array(seuils))
evolution = []
stop = False
while not stop:
    nb = 0
    for k in range(26):
        config = list(np.array(alphabet[k]).flatten())
        #nombre de modif total
        nbr = 0
        for i in range(len(config)):
            #indice de poids probable à être modifié 
            modif = []
            somme = 0
            n = 0
            for j in range(len(config)):
                if i != j and config[j] != False:
                    somme += weight[i][j]
                    modif.append(j)
                    n += 1
            somme -= seuil[i]
            #erreur 
            erreur = somme
            result = f_activation(somme)
            if result == int(config[i]):
                i += 1
            else:
                nbr += 1
                #delta erreur par poids
                if erreur > 0:
                    delta = (erreur + epsilon) / (n + 1)
                else:
                    delta = (erreur - epsilon) / (n + 1)
                #modification des poids
                for indice in modif:
                    weight[i][indice] -= delta
                    weight[indice][i] -= delta
                seuil[i] += delta
                #puis reprendre à zero avec le nouveau poids
                i = 0
        
        nb += nbr
    evolution.append(nb)
    #si le nombre d'erreur est nul alors on stop le programme
    if nb == 0:
        stop = True
print(f"nombre d'erreurs = {evolution}")
```

```python
plt.plot(evolution)
plt.bar([i for i in range(len(evolution))], evolution)
plt.ylabel('nombre d\'erreur')
plt.title('Evolution du nombre d\'erreur')
plt.show()
```
